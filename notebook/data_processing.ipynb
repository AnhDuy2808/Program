```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import joblib
```

```python
df = pd.read_csv('../data/pd_speech_features_processed.csv')

id_col = df['id'] if 'id' in df.columns else None
class_col = df['class'] if 'class' in df.columns else None

features = df.drop(columns=['id', 'class'], errors='ignore')


scaler = MinMaxScaler(clip=True)
features_scaled = scaler.fit_transform(features)

print(features_scaled)

df_scaled = pd.DataFrame(features_scaled, columns=features.columns)
if id_col is not None:
    df_scaled.insert(0, 'id', id_col)

if class_col is not None:
    df_scaled['class'] = class_col

df_scaled.to_csv('../data/pd_speech_features_scaled.csv', index=False)

joblib.dump(scaler, "../data/scaler/scaler.pkl")

features
```
